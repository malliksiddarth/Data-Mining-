```python
import pandas as pd

from decimal import Decimal
from math import sqrt, pow
import random
```

```python
def Euclidean_dis(x, y):  # Calculating Euclidean distance
    t = 0
    for i in range(len(x)):
        t = t + pow(Decimal(y[i]) - Decimal(x[i]), 2)
    euclidean_distance = sqrt(t)
    return round(euclidean_distance, 4)


def kmeans(kno, data):  # kmeans clustering
    centroids_n = [[random.uniform(1, 10) for i in range(7)] for i in range(kno)]
    sse = 0
    for n in range(100):
        centroids_info_hash_table = []
        sse_old = sse
        for i in range(len(data)):
            least_distance = min([Euclidean_dis(data[i], centroids_n[j]) for j in range(len(centroids_n))])
            for j in range(len(centroids_n)):
                if least_distance == Euclidean_dis(data[i], centroids_n[j]):
                    centroids_info_hash_table.append([j, least_distance])
        sse = sum([pow(i[1], 2) for i in centroids_info_hash_table])
        for i in range(len(centroids_n)):
            cnt = 0
            total = [0 for x in range(7)]
            for j in range(len(centroids_info_hash_table)):
                if centroids_info_hash_table[j][0] == i:
                    total = [total[k] + Decimal(data[j][k]) for k in range(len(total))]
                    cnt += 1
            if cnt != 0:
                centroids_n[i] = [x / cnt for x in total]
        if abs(sse - sse_old) < 0.0001:
            print(f'No of iterations = {n} SSE={sse}')
            return sse
    return sse


temp = []
with open(r'/Users/malliksiddarth/Desktop/FSU/Codes/Data Mining /Assignment 4/Data for Assignment 4/Data for Problem 2/seeds.txt') as f:
    temp = [i.split('   ') for i in f.read().splitlines()]
    for i in range(len(temp)):
        temp[i] = [j.strip() for j in temp[i] if j.strip() != '']

total = 0
for i in range(10):
    SSE = kmeans(kno=3, data=temp)
    total = total + SSE

print(f'Total average SSE for k = 3 =', total / 10)

total = 0
for i in range(10):
    SSE = kmeans(kno=5, data=temp)
    total = total + SSE

print(f'Total average SSE for k = 5 =', total / 10)

total = 0
for i in range(10):
    SSE = kmeans(kno=7, data=temp)
    total = total + SSE

print(f'Total average SSE for k = 7 =', total / 10)
```
